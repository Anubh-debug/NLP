```python
f = open('train_data.txt','r')
lines = f.readlines()
all_inletr = []
all_cletr = []
initial_Ltr = []
Lst_Ltr = []
```

```python
for line in lines:
  ln = line.split(',')
  if(ln[0]!='_'):
    initial_Ltr.append(ln[0])
    Lst_Ltr.append(ln[1].strip())
  else:
    if(len(initial_Ltr)>0 and len(Lst_Ltr)>0):
      all_inletr.append(initial_Ltr)
      all_cletr.append(Lst_Ltr)
    initial_Ltr = []
    Lst_Ltr = []
```

```python
data = {
      'raw letter' :all_inletr, 'correct letter': all_cletr 
  }
```

dividing both correct and incorrect words

```python
raw_data = pd.DataFrame(data)
raw_data
```

```python
letter_collection=[]
for word in all_inletr:
  for l in word:
    letter_collection.append(l)
letter_collection = list(set(letter_collection))
```

```python
df_emis = pd.DataFrame(index = letter_collection, columns=letter_collection)
df_emis = df_emis.fillna(0)
df_emis
```

```python
for i in range(len(raw_data)):
  rd = raw_data.iloc[i]
  #print(rd[0])
  n = len(rd[0])
  
  crc = rd[1]
  inc = rd[0]

  for j in range(n):
    df_emis.loc[crc[j],inc[j]]+=1
```

### Calculating emission probabilities

```python
nt = len(df_emis.iloc[:,0])
for i in range(nt):
  s = df_emis.iloc[i,:].sum()
  df_emis.iloc[i,:]/=s
df_emis
```

```python
n = len(raw_data)
m = len(letter_collection)
start_prob = [0]*m
freq=0
for l in letter_collection:
  rd = raw_data.loc[i]
  
  for i in range(n):
    rd = raw_data.loc[i]
    crc = rd[1]
    if(len(crc)>0 and crc[0]==l):
      freq+=1
  ind = letter_collection.index(l)
  start_prob[ind] = (freq/(n-1))
  freq=0
```

```python
for i in range(m):
  print(letter_collection[i] + " : " + str(start_prob[i]))
```

## Calculating transitional probabilities

```python
df_Trans = pd.DataFrame(index = letter_collection,columns = letter_collection)
df_Trans = df_Trans.fillna(0)
```

```python
nn = len(raw_data)
for i in range(nn):
  rd = raw_data.iloc[i]
  tset = rd[1]
  for j in range(1,len(tset)):
    df_Trans.loc[tset[j-1],tset[j]]+=1
df_Trans
```

```python
nt = len(df_Trans.iloc[:,0])
for i in range(nt):
  s = df_Trans.iloc[i,:].sum()
  df_Trans.iloc[i,:]/=s
df_Trans
```

```python
def Viterbi(word):
    state = []
    
     
    for key, letter in enumerate(word):
        p = [] 
        for tag in letter_collection:
            ind = letter_collection.index(tag)
            if key == 0:
                transition_p = start_prob[ind]
            else:
                transition_p = df_Trans.loc[state[-1], tag]
                 
            emission_p = df_emis.loc[tag,letter]
            state_probability = emission_p*transition_p    
            p.append(state_probability)
             
        pmax = max(p)
        state_max = letter_collection[p.index(pmax)] 
        state.append(state_max)
    return list(zip(word, state))
```

```python
p1 = "star wars is ploying at thi regal lloyd center and imax multnomah st portland ang also at tho centupy eastport plaza wuuld any of thoss times wurk for yoz"
given_sent = p1.split(" ")
predicted = []
t = []
for word in given_sent:
    r = []
    t = Viterbi(word)
    for tup in t:
      r.append(tup[1])
    predicted.append(r)
n_Sentence = ''
for i in predicted:
  s = ''
  for x in i:
    s = s + x
  n_Sentence = n_Sentence + s + ' '

n_Sentence
```

```python
f = open('test_data.txt','r')
lines = f.readlines()
all_inletr = []
all_cletr = []
initlLtr = []
cletr = []
for line in lines:
  ln = line.split(',')
  if(ln[0]!='_'):
    initlLtr.append(ln[0])
    cletr.append(ln[1].strip())
  else:
    if(len(initlLtr)>0 and len(cletr)>0):
      all_inletr.append(initlLtr)
      all_cletr.append(cletr)
    initlLtr = []
    cletr = []

print(all_inletr)
print(all_cletr)
```

```python
data = {
      'incorrect letter' :all_inletr, 'correct letter': all_cletr 
  }
test_set = pd.DataFrame(data)
test_set
```

Finding accuracy of model on test set

```python
tsz = len(test_set)
count=0
total=0
reslt_L = []
match_L = []
for i in range(tsz):
  match = []
  rt = test_set.iloc[i]
  ttag = rt[1]
  tword = rt[0]
  sn = len(tword)
  for j in range(sn):
    tp = (tword[j],ttag[j])
    match.append(tp)
  match_L.append(match)     # Storing each sentence word-tag tuple for test set in list
  reslt = Viterbi(tword)
  reslt_L.append(reslt)     # Storing each sentence word-tag tuple for return output set in list
  check = [i for i, j in zip(reslt, match) if i == j]
  count = count + len(check)
  total = total + len(reslt)

accuracy = count/total*100     
print("Overall accuracy: "+str(accuracy))        # Printing Overall Accuracy
```

```python
lt = len(match_L)
ltg = len(letter_collection)
accur = [0]*ltg
t_count=0
tg_c=0

for i in range(ltg):
  TP=0
  TN=0
  FP=0
  FN = 0
  precision = 0
  recall=0
  F_Score=0
  tag = list(letter_collection)[i]
  ind = list(letter_collection).index(tag)
  for j in range(lt):
    r_list = reslt_L[j]
    m_list = match_L[j]
    inL= len(r_list)
    for k in range(inL):
      if(m_list[k][1]==tag):
        t_count+=1
        if(r_list[k][1]==tag):
          tg_c+=1
          TP+=1
        else:
          FN+=1
      else:
        if(r_list[k][1]==tag):
          FP+=1
        else:
          TN+=1
  if(t_count!=0):
    accur[ind] = tg_c/t_count*100
  if (TP+FN)!=0:
    recall = TP/(TP+FN)
  if (TP+FP)!=0:
    precision = TP/(TP+FP)
  if((recall + precision)!=0):
    F_Score = 2* (recall*precision)/(recall + precision)
  print("For Class {}: Precision: {}, Recall: {}, F Score: {}".format(tag,precision,recall,F_Score))
  t_count=0
  tg_c=0

print("Class Wise Accuracy")
for i in range(ltg):
  print(list(letter_collection)[i] + ": " +str(accur[i]))
```
